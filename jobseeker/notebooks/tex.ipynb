```python
import subprocess

def generate_pdf(tex_file):
    # Command to compile the .tex file into a PDF
    command = ['pdflatex', '-interaction=nonstopmode', tex_file]
    
    try:
        # Run the command
        subprocess.run(command, check=True)
        print(f"PDF generated successfully from {tex_file}")
    except subprocess.CalledProcessError as e:
        print(f"Failed to generate PDF from {tex_file}: {e}")

# get the current directory
import os
cwd = os.getcwd()
print(cwd)


# Path to your .tex file
tex_file = "../data/cv/main.tex"
generate_pdf(tex_file)

# move the generated pdf to the output folder
```
